```python
import pickle
import pandas as pd
```

```python
loaded_artifact = pickle.load(open('credit_risk_model.sav', 'rb'))

model = loaded_artifact['model']
scaler = loaded_artifact['scaler']
freq_encoding_maps = loaded_artifact['freq_encoding_maps']
feature_names = loaded_artifact['feature_names']
```

```python
input_data = {
    'person_age': 28,
    'person_income': 45000,
    'person_emp_length': 5,
    'loan_amnt': 12000,
    'loan_int_rate': 11.5,
    'loan_percent_income': 0.25,
    'cb_person_cred_hist_length': 4,
    'person_home_ownership': 'RENT',
    'loan_intent': 'EDUCATION',
    'loan_grade': 'B',
    'cb_person_default_on_file': 'N',
    'age_group': '26-35',
    'income_group': 'low-middle',
    'loan_amount_group': 'large',
    'loan_to_income_ratio': 12000 / 45000,
    'loan_to_emp_length_ratio': 5 / 12000,
    'int_rate_to_loan_amt_ratio': 11.5 / 12000
}
```

```python
input_df = pd.DataFrame([input_data])
```

```python
for col, mapping in freq_encoding_maps.items():                            #Frequency Encoding
    input_df[col] = input_df[col].astype(str).map(mapping).fillna(0)
    
num_cols = scaler.feature_names_in_                                        #Scaling      
input_df[num_cols] = scaler.transform(input_df[num_cols])

input_df = input_df[feature_names]
```

```python
result = model.predict(input_df)
probability = model.predict_proba(input_df)[:, 1]

print("Prediction:", result[0])
print("Default Probability:", probability[0])
```
